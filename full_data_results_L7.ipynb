```python
import pandas as pd
import matplotlib.pyplot as plt
import prsa
from scipy.stats import ttest_rel
```

```python
RESULTS_FOLDER = 'results_L7'
```

```python
dfc = pd.read_csv(f'{RESULTS_FOLDER}/AC_DC_Values.csv', sep=';')
dfc.head()
```

```python
prsa.compare_capacities('Capacities', dfc['DC_normo'], dfc['AC_normo'], dfc['DC_hyper'], dfc['AC_hyper'])
```

```python
dfw = pd.read_csv(f'{RESULTS_FOLDER}/AC_DC_Windows.csv', sep=';')
dfw.head()
```

```python
prsa.compare_capacities('Windows', dfw['DC_normo_W'], dfw['AC_normo_W'], dfw['DC_hyper_W'], dfw['AC_hyper_W'])
```

```python
# Box plot for normocapnia and hypercapnia in deccerelation windows
plt.figure(figsize=(10, 6))
plt.boxplot([dfw['DC_normo_W'], dfw['DC_hyper_W']], labels=['Normokapnia', 'Hiperkapnia'])
plt.ylabel('')
plt.title('Windows for deceleration')
plt.show()

# Box plot for normocapnia and hypercapnia in deccerelation


plt.figure(figsize=(10, 6))
plt.boxplot([dfc['DC_normo'], dfc['DC_hyper']], labels=['Normokapnia', 'Hiperkapnia'])
plt.ylabel('ms')
plt.title('PRSA results for deceleration')
plt.show()
```

```python
# Box plot for normocapnia and hypercapnia in acceleration windows
plt.figure(figsize=(10, 6))
plt.boxplot([dfw['AC_normo_W'], dfw['AC_hyper_W']], labels=['Normokapnia', 'Hiperkapnia'])
plt.ylabel('')
plt.title('Windows for acceleration')
plt.show()

# Box plot for normocapnia and hypercapnia in aeccerelation


plt.figure(figsize=(10, 6))
plt.boxplot([dfc['AC_normo'], dfc['AC_hyper']], labels=['Normokapnia', 'Hiperkapnia'])
plt.ylabel('ms')
plt.title('PRSA results for acceleration')
plt.show()
```

```python
print(ttest_rel(dfc['AC_normo'], dfc['AC_hyper']))
print(ttest_rel(dfc['DC_normo'], dfc['DC_hyper']))
```
