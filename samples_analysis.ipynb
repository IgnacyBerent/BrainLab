```python
import os
import prep
import prsa
import matplotlib.pyplot as plt
```

```python
ndp= 'data/normokapnia/Ochotnik'
hdp = 'data/hiperkapnia/Ochotnik'

results = {}
normo_dc_data = []
normo_ac_data = []
hiper_dc_data = []
hiper_ac_data = []
normo_dc_w = []
normo_ac_w = []
hiper_dc_w = []
hiper_ac_w = []
```

# Analysis of file 1 - first 45000 elements of the hipercapnia file

```python
file_nr = 1

normo_file_path = ndp + f'0{file_nr}_baseline.csv'
hiper_file_path = hdp + f'0{file_nr}_hiperkapnia.csv'
normo_data = prep.read_data(normo_file_path, 'abp_finger[mm_Hg]')
hiper_data = prep.read_data(hiper_file_path, 'abp_finger[mm_Hg]')

hiper_data = hiper_data[:45000]

normo_dc, normo_ac, norm_dc_window, norm_ac_window = prsa.calculate_rr_dc_ac(signal_df = normo_data, percentile = 0.8, distance = 50,show_plot= True, plot_title = f'Normocapnia {file_nr}')
hiper_dc, hiper_ac, hyp_dc_window, hyp_ac_window = prsa.calculate_rr_dc_ac(signal_df = hiper_data, percentile = 0.8, distance = 50,show_plot= True, plot_title = f'Hipercapnia {file_nr}')

print(f'Normocapnia -> AC: {normo_ac:.4f}, DC: {normo_dc:.4f}')
print(f'Windows: {norm_ac_window}, {norm_dc_window}')
print(f'Hipercapnia -> AC: {hiper_ac:.4f}, DC: {hiper_dc:.4f}')
print(f'Windows: {hyp_ac_window}, {hyp_dc_window}')
print('-----------------------------------------')

#saving results for table
normo_dc_data.append(normo_dc)
normo_ac_data.append(normo_ac)
hiper_dc_data.append(hiper_dc)
hiper_ac_data.append(hiper_ac)

normo_dc_w.append(norm_dc_window)
normo_ac_w.append(norm_ac_window)
hiper_dc_w.append(hyp_dc_window)
hiper_ac_w.append(hyp_ac_window)
```

```python
import numpy as np

def unc_a(x: list[float]) -> float:
    """
    Function to calculate uncertainty of type a
    :param x: list of values
    :return: uncertainty of type a
    """
    n = len(x)
    mean_v = np.mean(x)
    bot = n*(n-1)
    top = np.sum([(i - mean_v)**2 for i in x])
    unc_a =  np.sqrt(top/bot)
    return unc_a
```

```python
print_unc = lambda c: print(f'{unc_a(c):.4f}')
```

```python
print_unc(normo_ac_data)
print_unc(normo_dc_data)
print_unc(hiper_ac_data)
print_unc(hiper_dc_data)
```
