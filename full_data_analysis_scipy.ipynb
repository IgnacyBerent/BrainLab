```python
import prep
import prsa
import os
import numpy as np
```

```python
FOLDER_NAME = 'scipy_results' # folder to save results in .csv format
```

```python
NORMO_DIR = './data_2/normokapnia'
HYPER_DIR = './data_2/hiperkapnia'

ABP_COLUMN = 'abp_finger_mm_hg_[abp_finger_mm_Hg_]'
ALTERNATIVE_ABP_COLUMN = 'abp_finger[abp_finger]'
```

```python
normo_dc_data = {}
normo_ac_data = {}
hyper_dc_data = {}
hyper_ac_data = {}

normo_dc_w = {}
normo_ac_w = {}
hyper_dc_w = {}
hyper_ac_w = {}
```

```python
files_n = []
mean_abp_n = {}
mean_hr_n = {}

files_h = []
mean_abp_h = {}
mean_hr_h = {}
```

```python
for filename in os.listdir(NORMO_DIR):
    if os.path.isfile(os.path.join(NORMO_DIR, filename)):
        file_path = os.path.join(NORMO_DIR, filename)
        try:
            df = prep.read_data(file_path, ABP_COLUMN, sep=';')
        except KeyError:
            df = prep.read_data(file_path, ALTERNATIVE_ABP_COLUMN, sep=';')

        files_n.append(df)
        prep.describe_file(df, filename)
        
        dc, ac, dc_w, ac_w = prsa.calculate_rr_dc_ac(
                                signal_df = df, percentile = 0.70, distance = 100, show_plot=False, plot_title=None)

        filename = filename[-7:-5]
        normo_dc_data[filename] = dc
        normo_ac_data[filename] = ac
        normo_dc_w[filename] = dc_w
        normo_ac_w[filename] = ac_w

        signal = df['Values'].interpolate(method='linear')
        mean_abp = signal.mean()
        # remove nan and inf values
        signal = signal[signal != np.inf]
        signal = signal[signal != -np.inf]
        signal = signal.dropna()
        HR = prep.calculate_mean_HR(signal)

        mean_abp_n[filename] = mean_abp
        mean_hr_n[filename] = HR

prep.files_time_analysis(files_n)
```

```python
for filename in os.listdir(HYPER_DIR):
    if os.path.isfile(os.path.join(HYPER_DIR, filename)):
        file_path = os.path.join(HYPER_DIR, filename)
        try:
            df = prep.read_data(file_path, ABP_COLUMN, sep=';')
        except KeyError:
            df = prep.read_data(file_path, ALTERNATIVE_ABP_COLUMN, sep=';')

        files_h.append(df)
        prep.describe_file(df, filename)
        
        dc, ac, dc_w, ac_w = prsa.calculate_rr_dc_ac(
                                signal_df = df, percentile = 0.70, distance = 100, show_plot=False, plot_title=None)

        filename = filename[-7:-5]
        hyper_ac_data[filename] = ac
        hyper_dc_data[filename] = dc
        hyper_ac_w[filename] = ac_w
        hyper_dc_w[filename] = dc_w

        signal = df['Values'].interpolate(method='linear')
        mean_abp = signal.mean()
        # remove nan and inf values
        signal = signal[signal != np.inf]
        signal = signal[signal != -np.inf]
        signal = signal.dropna()
        HR = prep.calculate_mean_HR(signal)

        mean_abp_h[filename] = mean_abp
        mean_hr_h[filename] = HR

prep.files_time_analysis(files_h)
```

```python
prep.save_to_csv(
    file_tittle = 'AC_DC_Values',
    folder_name = FOLDER_NAME,
    AC_normo = normo_ac_data, 
    AC_hyper = hyper_ac_data,
    DC_normo = normo_dc_data, 
    DC_hyper = hyper_dc_data,
    )
```

```python
prep.save_to_csv(
    file_tittle = 'AC_DC_Windows',
    folder_name = FOLDER_NAME,
    AC_normo_W = normo_ac_w,
    AC_hyper_W = hyper_ac_w,
    DC_normo_W = normo_dc_w,
    DC_hyper_W = hyper_dc_w
    )
```

```python
prep.save_to_csv(
    file_tittle = 'ABP_HR_Mean',
    folder_name = 'abp_hr_stats',
    abp_n = mean_abp_n,
    abp_h = mean_abp_h,
    hr_n = mean_hr_n,
    hr_h = mean_hr_h
    )
```
